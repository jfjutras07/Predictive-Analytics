## 3.5-Limitations

| Limitation | Description | Impact on Analysis |
|------------|------------|------------------|
| Feature Transformation Skipped | No advanced feature transformations (e.g., log, polynomial, or binning) were applied. | Some patterns in non-linear relationships may not be captured; cluster separations might be less distinct. |
| Outliers Retained | Outliers were intentionally kept in numeric features (e.g., Credit Score, Estimated Savings). | Can skew cluster centroids and reduce interpretability of some segments. |
| Small Sample Size | Dataset contains only 200 customers. | Limits generalizability of results; statistical reliability and cluster robustness are constrained. |
| Lack of Prescriptive/Transactional Data | No real transactional, revenue, or behavior data for evaluating financial impact or ROI. | Financial impact analysis and actionable recommendations are hypothetical. |
| No Distribution Adjustment | Skewed or highly asymmetric features were not transformed. | May influence distance-based clustering algorithms and internal validity metrics. |
| Synthetic Data Extension | Dataset is a synthetic extension of the original Kaggle dataset. | Patterns may not perfectly reflect real-world customer behavior; external validity is limited. |
